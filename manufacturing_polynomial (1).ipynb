# Reading The Dataset
Using Pandas for reading csv dataset format and print 5 first dataset from readed data.

```python
df = pd.read_csv('manufacturing.csv')
print(df.head())
```

# Dividing The Dataset Into 2 Part.
From dataset is divided the features into 2 part for learning and target features that named by x and y.

```python
x = df.iloc[:, :-2].values
y = df.iloc[:, -1].values
print(x)
print(y)
```

# Dividing The Dataset. Again.
The dataset is going to divided into training and testing dataset for evaluating the model.


Using test_size for managing the size for test dataset.

 and don't forget to keep the random_state into zero.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

# Cross-Check the dataset
cross-checking the dataset for avoiding the inconsistency amount of training and testing dataset.

```python
print(x_train)
print(y_train)
print(x_test)
print(y_test)
print(x_train.shape)
print(y_train.shape)
print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

# Create Polynomial Regression Model


```python
# train the model with polynomial regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg = PolynomialFeatures(degree=4)
x_poly_train = poly_reg.fit_transform(x_train)
x_poly_test = poly_reg.transform(x_test)
lin_reg = LinearRegression()
lin_reg.fit(x_poly_train, y_train)
```

# Predicting The Polynomial Model.
We can make the prediction by using the Test Dataset that has been provided after the splitting dataset.

```python
y_pred_train = lin_reg.predict(x_poly_train)
y_pred_test = lin_reg.predict(x_poly_test)
```

```python
y_pred = lin_reg.predict(poly_reg.fit_transform(x_test))
np.set_printoptions(precision=2)
print(y_pred)
```
